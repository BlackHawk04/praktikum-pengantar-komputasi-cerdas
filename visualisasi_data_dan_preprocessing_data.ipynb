# Visualisasi Data

## Matplotlib

### Diketahui data set iris.csv. Gunakan Scatter Plot untuk mevisualisasikan data iris tersebut.

Scatter Plot (alias diagram sebar, grafik sebar) menggunakan titik-titik untuk mewakili nilai dua variabel numerik yang berbeda . Posisi tiap titik pada sumbu horizontal dan vertikal menunjukkan nilai untuk suatu titik data individual. Plot sebar digunakan untuk mengamati hubungan antar variabel.

```python
# Import library matplotlib untuk membuat visualisasi data
import matplotlib.pyplot as plt

# Import library pandas dengan alias pd untuk membaca dataset iris dari file CSV
import pandas as pd

# Membaca dataset iris dari file 'iris.csv' dan memberi nama kolom-kolomnya
iris = pd.read_csv('/content/iris.csv', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])

# Menampilkan lima baris pertama dari dataset iris
print(iris.head())

# Membuat sebuah figure dan axis menggunakan metode subplots() dari matplotlib
fig, ax = plt.subplots()

# Menyebar (scatter) nilai sepal_length terhadap sepal_width pada axis yang telah dibuat sebelumnya
ax.scatter(iris['sepal_length'], iris['sepal_width'])

# Memberi judul pada plot
ax.set_title('Iris Dataset')

# Memberi label pada sumbu x (sepal_length)
ax.set_xlabel('sepal_length')

# Memberi label pada sumbu y (sepal_width)
ax.set_ylabel('sepal_width')

# Menampilkan plot
plt.show()
```

### Plot scatter dengan dengan mewarnai setiap titik data menurut kelasnya akan memberi lebih banyak arti pada grafik. Buatlah program untuk memvisualisasi data iris dengan memberikan warna yang berbeda untuk kelas yang berbeda

```python
# Mengimpor library matplotlib.pyplot sebagai plt dan pandas sebagai pd
import matplotlib.pyplot as plt
import pandas as pd

# Membaca dataset iris dari file CSV dengan nama kolom yang ditentukan
iris = pd.read_csv('/content/iris.csv', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])

# Menampilkan lima baris pertama dari dataset iris
print(iris.head())

# Membuat dictionary warna untuk setiap kelas iris
colors = {'Setosa':'r', 'Versicolor':'g', 'Virginica':'b'}

# Membuat objek figure dan axis menggunakan matplotlib
fig, ax = plt.subplots()

# Menggambar setiap titik data pada scatter plot berdasarkan kelas dengan warna yang sesuai
for i in range(len(iris['sepal_length'])):
    ax.scatter(iris['sepal_length'][i], iris['sepal_width'][i], color=colors[iris['class'][i]])

# Mengatur judul dan label sumbu pada plot
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')

# Menampilkan plot
plt.show()
```

## Line Chart
### Buatlah program untuk memplot grafik petal length dan petal width dari data iris

```python
# Mengimpor modul matplotlib.pyplot sebagai plt untuk membuat visualisasi grafik
import matplotlib.pyplot as plt

# Mengimpor modul pandas sebagai pd untuk manipulasi data menggunakan DataFrame
import pandas as pd

# Membaca dataset iris dari file CSV dan memberikan nama kolom
iris = pd.read_csv('/content/iris.csv', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])

# Menampilkan lima baris pertama dari dataset iris
print(iris.head())

# Mengambil nama kolom kecuali kolom 'class'
columns = iris.columns.drop(['class'])

# Membuat data x untuk plotting dengan rentang sesuai jumlah baris dalam dataset iris
x_data = range(0, iris.shape[0])

# Membuat figur dan sumbu menggunakan matplotlib
fig, ax = plt.subplots()

# Menggambar setiap kolom pada sumbu
for column in columns:
    ax.plot(x_data, iris[column], label=column)

# Menetapkan judul dan menampilkan legenda pada grafik
ax.set_title('Iris Dataset')
ax.legend()

# Menampilkan grafik
plt.show()
```

## Bar Chart
### Buatlah program untuk memvisualisasikan data review wine dalam bentuk histogram dengan metode Bar Chart.

```python
# Import library matplotlib untuk membuat visualisasi data
import matplotlib.pyplot as plt

# Import library pandas untuk bekerja dengan data dalam bentuk DataFrame
import pandas as pd

# Membaca data dari file CSV ('winemag.csv') dan menggunakan kolom pertama sebagai indeks
wine_reviews = pd.read_csv('winemag.csv', index_col=0)

# Menampilkan lima baris pertama dari DataFrame wine_reviews
print(wine_reviews.head())

# Membuat sebuah figur dan sumbu (figure and axis) untuk visualisasi data
fig, ax = plt.subplots()

# Menghitung frekuensi kemunculan setiap kelas pada kolom 'points' dari DataFrame wine_reviews
data = wine_reviews['points'].value_counts()

# Mendapatkan data x dan y (points dan frekuensinya)
points = data.index
frequency = data.values

# Membuat diagram batang (bar chart) menggunakan data points dan frequency
ax.bar(points, frequency)

# Menetapkan judul dan label sumbu pada diagram batang
ax.set_title('Wine Review Scores')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')

# Menampilkan diagram batang
plt.show()
```

## Box plots

Box plots adalah metode grafis untuk menampilkan ringkasan lima angka. Kita dapat membuat plot kotak menggunakan metode seaborns sns.boxplot dan mengirimkannya data serta nama kolom x dan y.

### Contoh plot dengan seaborn

```python
# Mengimpor modul matplotlib.pyplot sebagai plt
import matplotlib.pyplot as plt

# Mengimpor modul pandas sebagai pd
import pandas as pd

# Mengimpor modul seaborn sebagai sns
import seaborn as sns

# Membaca file CSV ('winemag.csv') dan menetapkan kolom pertama sebagai indeks
wine_reviews = pd.read_csv('/content/winemag.csv', index_col=0)

# Menampilkan lima baris pertama dari dataframe wine_reviews
print(wine_reviews.head())

# Membuat dataframe baru (df) yang berisi data anggur dengan skor (points) >= 95 dan harga (price) < 1000
df = wine_reviews[(wine_reviews['points']>=95) & (wine_reviews['price']<1000)]

# Membuat boxplot menggunakan seaborn, dengan sumbu x='points' dan sumbu y='price' dari dataframe df
sns.boxplot(x='points', y='price', data=df)

# Menampilkan plot
plt.show()
```

# Data Preprocessing

### Buatlah program untuk melakukan rescale data pima-indians-diabetes.csv dalam skala 0 sampai 1 dengan MinMaxScaler

```python
# Import library yang diperlukan
import pandas
import scipy
import numpy
from sklearn.preprocessing import MinMaxScaler

# Mendefinisikan URL file CSV dan nama kolom
url = "pima-indians-diabetes.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']

# Membaca data dari file CSV ke dalam pandas DataFrame
dataframe = pandas.read_csv(url, names=names)

# Mengonversi DataFrame menjadi array NumPy
array = dataframe.values

# Memisahkan array menjadi komponen input (X) dan output (Y)
X = array[:, 0:8]
Y = array[:, 8]

# Menginisialisasi objek MinMaxScaler untuk normalisasi fitur input (X)
scaler = MinMaxScaler(feature_range=(0, 1))

# Melakukan normalisasi pada fitur input (X)
rescaledX = scaler.fit_transform(X)

# Mengatur opsi tampilan untuk mencetak array NumPy dengan presisi 3 desimal
numpy.set_printoptions(precision=3)

# Mencetak hasil transformasi lima baris pertama dari fitur input yang telah dinormalisasi
print(rescaledX[0:5, :])
```

### Buatlah program untuk melakukan rescale data pima-indians-diabetes.csv dalam skala 0 sampai 1 dengan z-score

### Buatlah program untuk melakukan diskritasi data pima-indians-diabetes.csv dalam bentuk biner

```python
# Mengimpor kelas Binarizer dari modul preprocessing dalam pustaka scikit-learn
from sklearn.preprocessing import Binarizer
# Mengimpor pustaka pandas untuk membaca dan memanipulasi data dalam format dataframe
import pandas
# Mengimpor pustaka numpy untuk operasi numerik
import numpy

# Menentukan URL file CSV yang berisi data Pima Indians Diabetes
url = "/content/pima-indians-diabetes.csv"
# Menentukan nama kolom untuk dataframe
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']

# Membaca file CSV dan membuat dataframe menggunakan pandas
dataframe = pandas.read_csv(url, names=names)
# Mengubah dataframe menjadi array numpy
array = dataframe.values

# Memisahkan array menjadi komponen input (X) dan output (Y)
X = array[:, 0:8]
Y = array[:, 8]

# Membuat objek Binarizer dengan threshold 0.0 dan melatihnya menggunakan data input (X)
binarizer = Binarizer(threshold=0.0).fit(X)
# Menggunakan Binarizer yang telah dilatih untuk mentransformasi data input menjadi data biner
binaryX = binarizer.transform(X)

# Menyetel opsi cetak numpy untuk menampilkan angka desimal dengan 3 digit di belakang koma
numpy.set_printoptions(precision=3)
# Mencetak lima baris pertama dari data yang telah ditransformasi
print(binaryX[0:5, :])
```
